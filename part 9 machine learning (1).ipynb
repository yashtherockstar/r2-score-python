```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/Yash/Downloads/Cars93.csv")
```

```python
df.isna().sum()
```

```python
for i in df.columns:
    if df[i].dtypes==object:
        df[i].fillna(df[i].mode()[0],inplace=True)
    else:
        df[i].fillna(df[i].mean(),inplace=True)
```

```python
x=df.drop(labels=['Horsepower'],axis=1)
y=df[['Horsepower']]
```

```python
cat=[]
con=[]
for i in x.columns:
    if x[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
xcat=x[cat]
xcon=x[con]
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le=LabelEncoder()
ss=StandardScaler()
```

```python
for i in xcat.columns:
    xcat[i]=le.fit_transform(xcat[i])
```

```python
xcon=pd.DataFrame(ss.fit_transform(xcon),columns=con)
```

```python
x=xcon.join(xcat)
```

```python
out=[]
for i in xcon.columns:
    a=xcon[(xcon[i]<-3)|(xcon[i]>3)].index
    out.extend(a)
    
out
```

```python
outliers=list(set(out))
```

```python
outliers
```

```python
x.drop(index=outliers,axis=0,inplace=True)
y.drop(index=outliers,axis=0,inplace=True)
```
